Смоделировать игру против лотерейного автомата типа "777". 

Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. 

При некоторых значениях числа игрок получает выигрыш (см. справа)

1. Выгодна ли игра игроку? 
2. Сколько в среднем приобретает или теряет игрок за одну игру?

- 777: 200 руб. 
- 999: 100 руб.
- 555: 50 руб.
- 333: 15 руб.
- 111: 10 руб.
- *77: 5 руб.
- **7: 3 руб.
- *00: 2 руб.
- **0: 1 руб.

```python
import numpy as np
```

```python
np.random.randint(0, 999, 100)
```

#### Формула для вычисления среднего значения: X = P1X1 + P2X2 + ... 
Выпадение одного числа из 1000 возможных равно 0,001

Веротность выпадения *77 и *00 равно 0,045 и 0,02

Вероятность выпадения **7 равна 0,27

вероятность выпадения **0 равна 0,09

Тогда для 777 = 200 * 0,001 и так далее

```python
x = 0.2 + 0.1 + 0.05 + 0.015 + 0.01 + 0.045 + 0.27 + 0.02 + 0.09
x
```

Если я правильно понимаю, то данный результат говорит о невыгодности игры.
Так как на каждый вложенный рубль игрок возвращает 80 копеек

Проверим

```python
# np.random.seed(None)
trials = 10000    # Количество попыток
successes = 0    # Количество успехов
bingo = 1000 #количество рублей в начале игры
list_77 = [77, 177, 277, 377, 477, 577, 677, 877, 977]
list_00 = [100, 200, 300, 400, 500, 600, 700, 800, 900]
#list_7 = 

for i in range(trials):
    random_score = np.random.randint(0, 999)    # Случайное количество очков от 1 до 6
    bingo -= 1 # каждая игра стоит 1 рубль. Поэтому сразу его вычитаем
    if random_score == 777:
        successes += 1
        bingo += 200
        print('777')
        
    elif random_score == 999:
        successes += 1
        bingo += 100
        print('999')
        
    elif random_score == 555:
        successes += 1
        bingo += 50
        print('555')
        
    elif random_score == 333:
        successes += 1
        bingo += 15
        print('333')
        
    elif random_score == 111:
        successes += 1
        bingo += 10
        print('111')
        
    elif random_score in list_77:
        successes += 1
        bingo += 5
        print('*77')
        
    elif random_score % 10 == 7:
        successes += 1
        bingo += 3
        print('**7')
    
    elif random_score in list_00:
        successes += 1
        bingo += 2
        print('*00')
    
    elif random_score % 10 == 0:
        successes += 1
        bingo += 1
        print('**0')
    
    #else:
        #successes -= 1
     #   bingo -= 1
        
        
p_monte_carlo = successes / trials    # Вероятность из моделирования
p = 1 / 1000                             # Теоретическая вероятность

print(successes, trials, p_monte_carlo, p)
print('Было 1000 рубллей. Осталось', bingo)
```

на короткой дистанции кажется что есть шанс выиграть, но чем больше попыток, тем больше потери
